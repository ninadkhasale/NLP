```python
import numpy as np
import pandas as pd
from thefuzz import fuzz
from thefuzz import process
```

```python
df=pd.read_csv('base_names.csv')
df1=pd.read_csv('name_variations.csv')
```

```python
df
```

```python
df1
```

```python

# Matching variations with base names using fuzz.partial_ratio
matched_results = []

# Extract base names list
base_names = df['Base_Name'].tolist()

# Iterate over each variation and find the best match using fuzz.partial_ratio
for variation in df1['Variation']:
    match = process.extractOne(variation, base_names, scorer=fuzz.partial_ratio)
    matched_results.append(match[0])  # Appending variation and the best matched base name


# Convert results to DataFrame append it in the df1 
df1["match"]=matched_results

df1
```

```python
#accuracy
sum = np.sum(df1['match'] == df1['Matches_With_Base_Name'])
```

```python
#accuracy
accuracy = sum/len(df1) * 100

accuracy
```
